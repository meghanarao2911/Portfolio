```python
#check the percentage of null values in each of the columns and drop the columns that have max null values
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
#To remove highly correlated values
#function to remove higly correlated variables
def correlation(data, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i, j] >= threshold:
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in data.columns:
                    del data[colname] # deleting the column from the dataset

    return(data)
```

```python
#Remove highly negatively correlated values
def negative_correlation(data, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if corr_matrix.iloc[i, j] <= threshold:
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in data.columns:
                    del data[colname] # deleting the column from the dataset

    return(data)
```

```python
#function for outlier treatment
def remove_outliers(x):
    quant_df = list(x.quantile([0.95, 0.05]))
    Quant_3_1 = list(x.quantile([0.75, 0.25]))
    IQR = Quant_3_1[0] - Quant_3_1[1]
    IQR_LL = Quant_3_1[1] - 1.5*IQR
    IQR_UL = Quant_3_1[0] + 1.5*IQR
    x.loc[x > IQR_UL] = quant_df[0]
    x.loc[x < IQR_LL] = quant_df[1]
    return x  
```

```python
columns_for_outlier_treatment=["int_rate","annual_inc","dti","delinq_2yrs",
                               "inq_last_6mths","open_acc","pub_rec",
                               "revol_bal","revol_util",
                               "total_acc","out_prncp",
                               "total_pymnt","total_rec_int",
                               "total_rec_late_fee",
                               "recoveries",
                               "last_pymnt_amnt","collections_12_mths_ex_med","acc_now_delinq","tot_coll_amt","tot_cur_bal"]
```

```python
for cols in columns_for_outlier_treatment:
    data[cols] =remove_outliers(data[cols])
```
